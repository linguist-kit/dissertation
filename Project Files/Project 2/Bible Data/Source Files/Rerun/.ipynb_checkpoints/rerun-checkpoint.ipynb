```python
import pandas as pd
import spacy
from pathlib import Path

# Load the same ENG NT file you use downstream
eng_path = Path("eng_bib.txt")

# Load model
nlp = spacy.load("en_core_web_sm")

records = []
with open(eng_path, "r", encoding="utf-8") as f:
    for line in f:
        line = line.strip()
        if not line or line.startswith("#"):
            continue
        parts = line.split(maxsplit=1)
        if len(parts) != 2:
            continue
        vid, text = parts
        doc = nlp(text)
        for i, tok in enumerate(doc):
            if tok.pos_ == "AUX" and tok.lemma_ == "have":
                for j in range(i + 1, len(doc)):
                    if doc[j].pos_ in {"VERB", "AUX"} and doc[j].tag_ == "VBN":
                        records.append({"verse_id": vid, "english_text": text})
                        break
                break  # check only first HAVE per verse

df_plg = pd.DataFrame(records)
df_plg.to_csv("plg_verses_new.csv", index=False, encoding="utf-8")
print(f"✅ Saved {len(df_plg)} verses to plg_verses_new.csv")
```
