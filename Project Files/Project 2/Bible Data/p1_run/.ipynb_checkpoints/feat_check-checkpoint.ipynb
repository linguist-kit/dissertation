```python
# === Run to check for and report on feature mismatches between human- and machine-annotated datasets ===

import pandas as pd

# --- paths ---
base = "/Users/glossophilia/Library/CloudStorage/GoogleDrive-rumilofaniel@gmail.com/My Drive/Diss/Project Files/Project 1/Bible Data/p1_run"

ha_path = f"{base}/p1_run_ha.csv"
ag_path = f"{base}/p1_run_ag.csv"

# --- load ---
df_ha = pd.read_csv(ha_path)
df_ag = pd.read_csv(ag_path)

# Strip empty rows
df_ha = df_ha[df_ha["TokenID"].astype(str).str.strip() != ""].reset_index(drop=True)
df_ag = df_ag[df_ag["TokenID"].astype(str).str.strip() != ""].reset_index(drop=True)

# --- define feature columns explicitly ---
feature_cols = [
    "DtExp", "DtRes", "DtUniv", "DtDTAM",
    "TLTS", "TLAspPfv", "TLMdIrr",
    "ClNegPol", "ClCntfcl", "VcPass",
    "ASTrns", "ASUnrg", "ASUncc",
    "AktState", "AktAct", "AktAcc", "AktAch"
]

# --- align on TokenID ---
ha = df_ha[["TokenID"] + feature_cols].set_index("TokenID")
ag = df_ag[["TokenID"] + feature_cols].set_index("TokenID")

# sanity check (optional but comforting)
assert ha.index.equals(ag.index), "TokenIDs do not align!"

# --- find mismatches ---
rows = []

for feat in feature_cols:
    mismatch = ha[feat] != ag[feat]
    mismatch = mismatch & ~(ha[feat].isna() & ag[feat].isna())

    for token_id in ha.index[mismatch]:
        rows.append({
            "TokenID": token_id,
            "Feature": feat,
            "HA": ha.at[token_id, feat],
            "AG": ag.at[token_id, feat],
        })

diff_df = pd.DataFrame(rows)

print(f"Total mismatches: {len(diff_df)}")
diff_df.head()
```

```python
# === Create reports ===

# Optional polish: sort for readability
diff_df = diff_df.sort_values(["Feature", "TokenID"])

out_path = f"{base}/p1_feature_mismatches.csv"
diff_df.to_csv(out_path, index=False)
print("Saved to:", out_path)

out_path = f"{base}/p1_feature_mismatches.txt"
with open(out_path, "w") as f:
    for _, row in diff_df.iterrows():
        f.write(f"{row.TokenID}\t{row.Feature}\tHA={row.HA}\tAG={row.AG}\n")

print("Saved to:", out_path)
```
