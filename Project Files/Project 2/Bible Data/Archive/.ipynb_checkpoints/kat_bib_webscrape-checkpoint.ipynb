```python
#Import packages

import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#Start with Genesis Chapter 1

#Define the target URL for Genesis 1
url = "https://holybible.ge/geo/bible/სბს–2013/1/0/0/0/1" #"https://holybible.ge/geo/bible/%E1%83%A1%E1%83%91%E1%83%A1%E2%80%932013/1/0/0/0/1"

#Add headers to mimic a real browser
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0 Safari/537.36"
}

#Fetch the page using headers
response = requests.get(url, headers=headers)

#Check for access issues
if response.status_code != 200:
    print("Access blocked. Status code:", response.status_code)
else:
    #Parse page
    soup = BeautifulSoup(response.content, "html.parser")

    #Extract all visible text
    text = soup.get_text()

    #Clean up the text
    lines = [line.strip() for line in text.split("\n") if line.strip()]

    #Preview some of the lines
    for line in lines[:20]:
        print(line)
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import time

# Setup visible browser (for debugging)
options = Options()
# options.add_argument("--headless")  # REMOVE this for now
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

# Launch browser
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# Target URL
url = "https://holybible.ge/geo/bible/სბს–2013/1/0/0/0/1"
driver.get(url)

# Wait for page to load JS
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Wait for the verse elements to be present
WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.CLASS_NAME, "versetext"))
)

# Parse loaded HTML
soup = BeautifulSoup(driver.page_source, "html.parser")

# Find all verse <div> blocks
verses = soup.find_all("div", class_="verse")

verse_data = []

for verse in verses:
    verse_number_tag = verse.find("span", class_="versenumber")
    verse_text_tag = verse.find("span", class_="versetext")

    if verse_number_tag and verse_text_tag:
        verse_number = verse_number_tag.text.strip()
        text = verse_text_tag.text.strip()

        verse_data.append((verse_number, text))
    
    text = verse.text.strip()
    
    # Remove the number prefix from the text
    if text.startswith(verse_number):
        text = text[len(verse_number):].strip()
    
    verse_data.append((verse_number, text))

# Convert to DataFrame for easy handling
df = pd.DataFrame(verse_data, columns=["Verse", "Text"])
df.index += 1  # Start row numbers at 1 for readability

# Preview
df.head()

# Extract all visible text
text = soup.get_text()
lines = [line.strip() for line in text.split("\n") if line.strip()]

# Preview
for line in lines[:20]:
    print(line)

# Close the browser
driver.quit()
```
