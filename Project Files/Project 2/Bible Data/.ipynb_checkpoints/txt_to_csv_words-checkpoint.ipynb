```python
#Takes a .txt Bible file and converts it to a word-seprated .csv file
```

```python
import csv


input_path = "eng-x-bible-newsimplified.txt"   # change to your input file
output_path = "eng_source.csv"  # output CSV
language_code = "ENG"              # change to DEU, FRA, etc.

with open(input_path, "r", encoding="utf-8") as infile, \
     open(output_path, "w", newline="", encoding="utf-8") as outfile:
    
    writer = csv.writer(outfile)
    
    for line in infile:
        line = line.strip()
        if not line:
            continue
        
        # Split only once to separate verse ID from text
        verse_id, text = line.split(maxsplit=1)
        
        # Split text into words
        words = text.split()
        
        # Write the row: verse ID, language code, words
        writer.writerow([verse_id, language_code] + words)

print(f"✅ Done! Created {output_path}")
```
