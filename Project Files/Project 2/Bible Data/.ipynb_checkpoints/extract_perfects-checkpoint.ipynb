```python
#Takes a .txt (in this case bilingual ENG+LANG) file as input and returns a list of verse numbers
```

```python
# Step 1: Extract English verse lines from the glossed file

from pathlib import Path
import pandas as pd

# Change this path to point to your glossed file
file_path = Path("deu-x-bible-zuercher_glossed.txt")

verse_id = None
records = []

with open(file_path, "r", encoding="utf-8") as f:
    for line in f:
        line = line.strip()
        if not line:
            continue

        # If this line starts a new verse block
        if line.startswith("verse:"):
            verse_id = line.replace("verse:", "").strip()

        # If this line contains the English text for that verse
        elif line.startswith("ENG:") and verse_id is not None:
            text = line.replace("ENG:", "").strip()
            records.append({"verse_id": verse_id, "english_text": text})

# Convert to DataFrame
df_eng = pd.DataFrame(records)

print(f"✅ Extracted {len(df_eng)} English verses")
display(df_eng.head(10))
```

```python
import spacy

# Load English spaCy model
nlp = spacy.load("en_core_web_sm")

# Store matches here
matches = []

for _, row in df_eng.iterrows():
    verse_id = row["verse_id"]
    text = row["english_text"]

    doc = nlp(text)

    for i, token in enumerate(doc):
        # Look for "have" as auxiliary
        if token.pos_ == "AUX" and token.lemma_ == "have":
            # Look ahead for a past participle verb (VBN)
            for j in range(i + 1, len(doc)):
                if doc[j].pos_ in {"VERB", "AUX"} and doc[j].tag_ == "VBN":
                    matches.append(verse_id)
                    break
            break  # only check first HAVE per sentence

# Create DataFrame of matches
df_matches = df_eng[df_eng["verse_id"].isin(matches)].reset_index(drop=True)

print(f"✅ Found {len(df_matches)} verses with 'have + VBN' pattern")
display(df_matches.head(10))
```

```python
# Step 3: Export verse IDs (and optionally English text) to CSV

output_path = "plg_verses.csv"

# Save only verse_id and English text
df_matches[["verse_id", "english_text"]].to_csv(output_path, index=False, encoding="utf-8")

print(f"✅ Saved {len(df_matches)} matching verses to {output_path}")
```
