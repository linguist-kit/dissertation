```python
from PyPDF2 import PdfReader

#Load the PDF
reader = PdfReader("NT_kat.pdf")

#Extract all text
all_text = ""
for page in reader.pages:
    all_text += page.extract_text() + "\n"

#Preview the first 1000 characters
print(all_text[:1000])
```

```python
#Structure the text into a csv

import pandas as pd

data = []

book = "მათე"  # Start with Matthew
chapter = 1
verse = 0

verse_pattern = re.compile(r'^(\d+)\s+(.*)')

for line in cleaned_lines:
    #Try to match "1 text..." or "23 text..."
    match = verse_pattern.match(line)
    if match:
        verse = int(match.group(1))
        text = match.group(2)
        data.append([book, chapter, verse, text])
    elif "თავი" in line: # Look for "თავი 2", "თავი 3" as chapter headers
        ch_match = re.search(r"თავი\s+(\d+)", line)
        if ch_match:
            chapter = int(ch_match.group(1))

#Save to csv
df = pd.DataFrame(data, columns=["Book", "Chapter", "Verse", "Text"])
df.to_csv("kat_nt_sample.csv", index=False, encoding="utf-8-sig")
```
