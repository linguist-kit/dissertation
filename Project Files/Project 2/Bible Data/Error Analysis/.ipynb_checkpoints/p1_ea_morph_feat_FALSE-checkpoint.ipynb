```python
import pandas as pd

# --- Paths (edit these) ---
PRED_PATH = "/Users/glossophilia/Library/CloudStorage/GoogleDrive-rumilofaniel@gmail.com/My Drive/Diss/Project Files/Project 1/Bible Data/Error Analysis/y_pred_rfc_long.csv"
OUT_PATH  = "/Users/glossophilia/Library/CloudStorage/GoogleDrive-rumilofaniel@gmail.com/My Drive/Diss/Project Files/Project 1/Bible Data/Error Analysis/errors_rfc_with_features.csv"

# --- Load predictions (already long format) ---
preds = pd.read_csv(PRED_PATH)

# Keep only errors
errors = preds[preds["correct"] == False].copy()

# --- Attach features/context ---
# Option A: if you still have df_proc in memory in this notebook, use it:
feature_cols = [
    "TokenID",
    "DtExp", "DtRes", "DtUniv", "DtDTAM",
    "TLTS", "TLAspPfv", "TLMdIrr",
    "ClNegPol", "ClIntgv", "ClCntfcl",
    "VcPass",
    "ASTrns", "ASUnrg", "ASUncc",
    "AktState", "AktAct", "AktAcc", "AktAch"
]
# Load dev features from CSV (exported from the pipeline)
X_DEV_PATH = "/Users/glossophilia/Library/CloudStorage/GoogleDrive-rumilofaniel@gmail.com/My Drive/Diss/Project Files/Project 1/Bible Data/Error Analysis/X_dev.csv"
X_dev = pd.read_csv(X_DEV_PATH)

features_df = X_dev[[c for c in feature_cols if c in X_dev.columns]].drop_duplicates("TokenID")

# Merge features onto errors
errors_feat = errors.merge(features_df, on="TokenID", how="left")

# Export
errors_feat.to_csv(OUT_PATH, index=False)

print("✅ Exported:", OUT_PATH)
print("Rows (errors):", len(errors_feat))
print("Missing feature rows:", errors_feat.filter(like="Dt").isna().all(axis=1).sum())
```

```python
def sample_errors(mismatches, n_per_pair=5, random_state=0):
    """
    Draw up to n_per_pair random examples from each (gold, pred) confusion pair.
    Helpful for quick manual review.
    """
    return (
        mismatches.groupby(["gold", "pred"], group_keys=False)
        .apply(lambda g: g.sample(min(n_per_pair, len(g)), random_state=random_state))
        .reset_index(drop=True)
    )

# Example usage after running build_mismatch_table:
sampled = sample_errors(mismatches, n_per_pair=3)
sampled.head()
```
