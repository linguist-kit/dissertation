```python
#1. Function

import pandas as pd

def add_verification_columns(df):
    yn = lambda cond: cond.map({True: 'Y', False: 'N'})

    # Perfect readings
    df['DtExp/DtRes']        = yn((df['DtExp']==1) & (df['DtRes']==1))
    df['DtExp/DtUniv']       = yn((df['DtExp']==1) & (df['DtUniv']==1))
    df['DtRes/DtUniv']       = yn((df['DtRes']==1) & (df['DtUniv']==1))
    df['DtExp/DtRes/DtUniv'] = yn((df['DtExp']==1) & (df['DtRes']==1) & (df['DtUniv']==1))
    df['DtHP']               = yn((df['DtExp']==0) & (df['DtRes']==0) & (df['DtUniv']==0))
    df['DtHP_excused']       = yn((df['DtHP']=='Y') & (df['TLMdIrr']==1))

    # Argument structure
    df['ASTrns/ASUncc']        = yn((df['ASTrns']==1) & (df['ASUncc']==1))
    df['ASTrns/ASUnrg']        = yn((df['ASTrns']==1) & (df['ASUnrg']==1))
    df['ASUncc/ASUnrg']        = yn((df['ASUncc']==1) & (df['ASUnrg']==1))
    df['ASTrns/ASUncc/ASUnrg'] = yn((df['ASTrns']==1) & (df['ASUncc']==1) & (df['ASUnrg']==1))
    df['ASHP']                 = yn((df['ASTrns']==0) & (df['ASUncc']==0) & (df['ASUnrg']==0))

    # Aktionsart
    akt_cols = ['AktState', 'AktAct', 'AktAcc', 'AktAch', 'AktSemel']
    df['AktState/AktAct'] = yn((df['AktState']==1) & (df['AktAct']==1))
    df['AktAcc/AktAch']   = yn((df['AktAcc']==1)   & (df['AktAch']==1))
    df['AktMultiple']     = yn(df[akt_cols].sum(axis=1) >= 2)
    df['AktHP']           = yn(df[akt_cols].sum(axis=1) == 0)

    # Unexpected perfective pairings
    df['TLAspPfv/DtUniv']  = yn((df['TLAspPfv']==1) & (df['DtUniv']==1))
    df['TLAspPfv/AktState']= yn((df['TLAspPfv']==1) & (df['AktState']==1))
    df['TLAspPfv/AktAct']  = yn((df['TLAspPfv']==1) & (df['AktAct']==1))

    return df
```

```python
from pathlib import Path

BASE_DIR   = Path("/Users/glossophilia/Library/CloudStorage/GoogleDrive-rumilofaniel@gmail.com/My Drive/Diss/Project Files/Project 1/Runs/OpenAI")
merged_dir = BASE_DIR

for csv_path in sorted(merged_dir.glob('*.csv')):
    if '_ver' in csv_path.stem:
        continue
    df = pd.read_csv(csv_path)
    df = add_verification_columns(df)
    out_path = csv_path.parent / (csv_path.stem + '_ver.csv')
    df.to_csv(out_path, index=False)
    print(f"Written: {out_path.name}")
```
