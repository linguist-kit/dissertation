```python
import pandas as pd
import os
import glob

BASE_DIR = "/Users/glossophilia/Library/CloudStorage/GoogleDrive-rumilofaniel@gmail.com/My Drive/Diss/Project Files/Project 1/Comps/Sources"
BASE_FILE = "KD_annotations_350.csv"
OUTPUT_DIR = BASE_DIR

COLS = [
    "ASTrns", "ASUncc", "ASUnrg",
    "AktAcc", "AktAch", "AktAct", "AktSemel",
    "AktState",
    "ClCntfcl", "ClNegPol",
    "DtDTAM", "DtExp", "DtRes", "DtUniv",
    "TLAspPfv", "TLMdIrr", "TLTS",
    "VcPass",
]
KEY_COL = "TokenID"

base = pd.read_csv(os.path.join(BASE_DIR, BASE_FILE)).set_index(KEY_COL)[COLS].sort_index()

compare_files = [
    f for prefix in ("Claude_", "OpenAI_")
    for f in glob.glob(os.path.join(BASE_DIR, f"{prefix}*_ver.csv"))
]

print(f"Found {len(compare_files)} files to compare.\n")

for comp_path in sorted(compare_files):
    comp_name = os.path.splitext(os.path.basename(comp_path))[0]
    version = comp_name.split('_', 1)[1]
    comp = pd.read_csv(comp_path).set_index(KEY_COL)[COLS].sort_index()

    shared = base.index.intersection(comp.index)
    if len(shared) < len(base.index):
        print(f"  Note: {len(base.index) - len(shared)} base row(s) and {len(comp.index) - len(shared)} comp row(s) skipped (no matching TokenID)")
    base_aligned = base.loc[shared]
    comp_aligned = comp.loc[shared]

    match = (base_aligned.astype(int) == comp_aligned).replace({True: "Y", False: "N"})
    base_renamed = base_aligned.rename(columns={c: f"{c}_KD" for c in COLS})
    comp_renamed = comp_aligned.rename(columns={c: f"{c}_{version}" for c in COLS})

    interleaved_cols = [col for c in COLS for col in (f"{c}_KD", f"{c}_{version}", c)]
    result = pd.concat([base_renamed, comp_renamed, match], axis=1)[interleaved_cols]

    out_name = comp_name.removesuffix("_ver") + "_comp.csv"
    result.reset_index().to_csv(os.path.join(OUTPUT_DIR, out_name), index=False)
    print(f"Done: {out_name}")
```
