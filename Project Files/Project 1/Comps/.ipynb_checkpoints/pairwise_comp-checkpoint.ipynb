```python
import pandas as pd
import os

BASE_DIR = "/Users/glossophilia/Library/CloudStorage/GoogleDrive-rumilofaniel@gmail.com/My Drive/Diss/Project Files/Project 1/Comps"
BASE_FILE = "Gemini_full_r.csv"
COMPARE_FILE = "KD_annotations_350.csv"
OUTPUT_DIR = "/Users/glossophilia/Library/CloudStorage/GoogleDrive-rumilofaniel@gmail.com/My Drive/Diss/Project Files/Project 1/Comps/Results"

COLS = [
    "ASTrns", "ASUncc", "ASUnrg",
    "AktAcc", "AktAch", "AktAct", "AktSemel",
    "AktState",
    "ClCntfcl", "ClNegPol",
    "DtDTAM", "DtExp", "DtRes", "DtUniv",
    "TLAspPfv", "TLMdIrr", "TLTS",
    "VcPass",
]
KEY_COL = "TokenID"

os.makedirs(OUTPUT_DIR, exist_ok=True)

base = pd.read_csv(os.path.join(BASE_DIR, BASE_FILE)).set_index(KEY_COL)[COLS].sort_index()
comp = pd.read_csv(os.path.join(BASE_DIR, COMPARE_FILE)).set_index(KEY_COL)[COLS].sort_index()

shared = base.index.intersection(comp.index)
if len(shared) < len(base.index) or len(shared) < len(comp.index):
    print(f"Note: {len(base.index) - len(shared)} base row(s) and {len(comp.index) - len(shared)} comp row(s) skipped (no matching TokenID)")
base_aligned = base.loc[shared]
comp_aligned = comp.loc[shared]

# NaN-aware comparison: NaN vs NaN → Y, NaN vs value → N
match = (base_aligned.eq(comp_aligned) | (base_aligned.isna() & comp_aligned.isna()))
match = match.replace({True: "Y", False: "N"})

# Derive column suffixes from filenames
base_sfx = os.path.splitext(BASE_FILE)[0]
comp_sfx = os.path.splitext(COMPARE_FILE)[0]

base_renamed = base_aligned.rename(columns={c: f"{c}_{base_sfx}" for c in COLS})
comp_renamed = comp_aligned.rename(columns={c: f"{c}_{comp_sfx}" for c in COLS})

interleaved_cols = [col for c in COLS for col in (f"{c}_{base_sfx}", f"{c}_{comp_sfx}", c)]
result = pd.concat([base_renamed, comp_renamed, match], axis=1)[interleaved_cols]

output_name = f"{base_sfx}_comp.csv"
result.reset_index().to_csv(os.path.join(OUTPUT_DIR, output_name), index=False)
print(f"Done: {output_name}")
```
