```python
from collections import defaultdict
from collections import Counter  # add this at the top
import string  # for checking punctuation

input_path = "swe-x-bible-folk1998_glossed.txt"

# Helper: define a set of punctuation characters
PUNCT = set(string.punctuation)

# Build a new dictionary
freq_dict = defaultdict(Counter)

with open(input_path, 'r', encoding='utf-8') as f:
    lines = f.readlines()

for i in range(0, len(lines), 4):
    lang_line = lines[i + 2].strip()
    tokens = lang_line.split()

    for token in tokens:
        if '/' in token:
            word, gloss = token.split('/', 1)

            # Filter out empty glosses
            if not gloss.strip():
                continue

            # Optional: filter if gloss is *only* punctuation
            if all(char in PUNCT for char in gloss.strip()):
                continue

            # Optional: filter if word is only punctuation
            if all(char in PUNCT for char in word.strip()):
                continue

            # Normalize gloss
            norm = normalize_gloss(gloss)
            if not norm:
                continue

            # Count it
            freq_dict[word][norm] += 1

# Preview filtered results
print("Filtered gloss dictionary entries (with frequencies):")
for i, (word, gloss_counter) in enumerate(freq_dict.items()):
    # Sort glosses by frequency, descending
    sorted_glosses = [f"{gloss} ({count})" for gloss, count in gloss_counter.most_common()]
    print(f"{word} → {', '.join(sorted_glosses)}")
    if i >= 20:
        break
```

```python
top_n_dict = {}

N = 3  # change this to 1 for top gloss only, or 3+ if you want more

for word, gloss_counter in freq_dict.items():
    top_glosses = [gloss for gloss, count in gloss_counter.most_common(N)]
    top_n_dict[word] = top_glosses

print("Top glosses per word:")
for i, (word, glosses) in enumerate(top_n_dict.items()):
    print(f"{word} → {', '.join(glosses)}")
    if i >= 20:
        break
```

```python
# This is just a safety pass if you want to re-normalize top_n_dict

normalized_top_n_dict = {}

for word, glosses in top_n_dict.items():
    cleaned_glosses = []
    for gloss in glosses:
        norm = normalize_gloss(gloss)
        if norm:
            cleaned_glosses.append(norm)
    if cleaned_glosses:
        normalized_top_n_dict[word] = cleaned_glosses

# Preview result
print("Sample normalized top gloss entries:")
for i, (word, glosses) in enumerate(normalized_top_n_dict.items()):
    print(f"{word} → {', '.join(sorted(glosses))}")
    if i >= 20:
        break
```
