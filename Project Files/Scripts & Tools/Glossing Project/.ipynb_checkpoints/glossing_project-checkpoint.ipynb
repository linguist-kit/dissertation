```python
# === MINI GLOSSING PROJECT ===
# Goal: align Hindi tokens with English glosses using simulated alignments

# Step 1: Sample sentence pair
hin_tokens = ["ibraaheem", "kee", "santan"]
eng_tokens = ["descendant", "of", "abraham"]

# Step 2: Simulated alignment from GIZA++ or fast_align
# Format: (eng_index, hin_index)
alignment = [(2, 0), (1, 1), (0, 2)]  # ibraaheem ↔ abraham, kee ↔ of, santan ↔ descendant

# Step 3: Build glossed output
glossed = []
for i, hin_word in enumerate(hin_tokens):
    # Find aligned English word
    gloss = ""
    for eng_idx, hin_idx in alignment:
        if hin_idx == i:
            gloss = eng_tokens[eng_idx].upper()
            break
    glossed.append(f"{hin_word}/{gloss}")

# Step 4: Display result
print("Hindi:", " ".join(hin_tokens))
print("English:", " ".join(eng_tokens))
print("Glossed:", " ".join(glossed))
```
