```python
def read_record_from_file(f, lang_code):
    record = {}
    keys = ['verse', 'ENG', lang_code]
    for key in keys:
        # Grab a line from the file and strip the trailing newline.
        line = next(f).strip()

        # Make sure this line really starts with this key.
        assert(line.startswith(key))

        # Split the key on the left from the content on the right.
        parts = line.split(maxsplit=1)
        
        # Make sure it actually has content on the right.
        assert(len(parts) == 2)

        record[key] = parts[1]

    # Grab one more line, which should be blank.
    line = next(f).strip()
    assert(line == '')

    return record
```

```python
def parse_bible_file(filename, lang_code):
    """Reads through bible file with ``filename`` and language code
    ``lang_code`` and returns a dictionary with keys containing
    the verse number (as a string) and values containing the
    glossed verse in the target language."""
    # Initialize an empty dict to store the data in.
    verse_to_gloss = {}
    with open(filename, 'r') as f:
        # Keep looping until we ``break``.
        while True:
            try:
                # Read one record at a time.
                record = read_record_from_file(f, lang_code)
                # Store the verse and glossed verse in key and the glossed verse
                # for ``lang_code`` in the value.
                verse_to_gloss[record['verse']] = record[lang_code]
            # If we hit the end of the file, exit this loop.
            except StopIteration:
                break
    # Return the dictionary ``verse_to_gloss``, which is now populated
    # with data we can use to look up a glossed verse by its verse number.
    return verse_to_gloss
```

```python
def verse_numbers_from_csv(filename):
    """Reads the CSV file with ``failname`` and returns a list
    of strings containing bible verse numbers in the TODO column."""
    verse_numbers = []
    with open(filename, 'r') as f:
        csv_reader = csv.DictReader(f)
        for record in csv_reader:
            verse_numbers.append(record['verse'])
    return verse_numbers
```
