```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
sentence = "I have read that book before."
embedding = model.encode(sentence)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Two sample sentences
sentence1 = "I have read that book before."
sentence2 = "I've already finished reading that book."

# Generate embeddings
embedding1 = model.encode(sentence1)
embedding2 = model.encode(sentence2)

# Compare using cosine similarity
similarity = cosine_similarity([embedding1], [embedding2])[0][0]
print(f"Cosine similarity: {similarity:.4f}")
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Two sample sentences
sentence1 = "I have eaten sushi before."
sentence2 = "At that point I had already tried Japanese food."

# Generate embeddings
embedding1 = model.encode(sentence1)
embedding2 = model.encode(sentence2)

# Compare using cosine similarity
similarity = cosine_similarity([embedding1], [embedding2])[0][0]
print(f"Cosine similarity: {similarity:.4f}")
```

```python
sentences = [
    "I have read that book before.",               # anchor
    "I've already finished reading that book.",    # similar
    "The cat sat on the mat.",                     # unrelated
    "He has never visited France."                 # kinda close (perfect-like)
]

# Generate embeddings
embeddings = model.encode(sentences)

# Compare anchor to each of the others
from sklearn.metrics.pairwise import cosine_similarity

anchor = embeddings[0].reshape(1, -1)
others = embeddings[1:]

similarities = cosine_similarity(anchor, others)[0]

# Print results
for i, score in enumerate(similarities, start=1):
    print(f"Similarity to sentence {i+1}: {score:.4f}")
```
