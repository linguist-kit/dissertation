```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
sentences = [
    "We go there.",
    "We go there often.",
    "We are going there.",
    "We will go there.",
    "We may go there.",
    "We went there."
    "We have gone there.",
    "We may have gone there.",
    "We would have gone there.",
    "We must go there."
]

# Generate embeddings
embeddings = model.encode(sentences)

# Compare anchor to each of the others
from sklearn.metrics.pairwise import cosine_similarity

anchor = embeddings[0].reshape(1, -1)
others = embeddings[1:]

similarities = cosine_similarity(anchor, others)[0]

# Print results
for i, score in enumerate(similarities, start=1):
    print(f"Anchor: {sentences[0]}")
    print(f"Compare: {sentences[i]}")
    print(f"Similarity: {score:.4f}")
    print("-" * 50)
```

```python
import pandas as pd

# Build dataframe of results
results_df = pd.DataFrame({
    "Anchor": [sentences[0]] * (len(sentences) - 1),
    "Compare": sentences[1:], 
    "Similarity": similarities
})

# Save to CSV
results_df.to_csv("TAM_contrasts_results.csv", index=False)
```
